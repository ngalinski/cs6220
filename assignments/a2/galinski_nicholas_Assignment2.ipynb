# Part 1: Iris Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
fileURL = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data" 
df = pd.read_csv(fileURL, names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Name"], header=None)
df
```

#### 1.1 Summary Statistics

```python
df.head()
```

```python
# number of features
len(df.columns)
```

```python
# data type of features
df.dtypes
```

```python
# summary statistics - minimum value, maximum value, mean, range, standard deviation, 
# variance, count, and 25:50:75% percentiles
df.describe(include = 'all')
```

```python
# variance isn't included above - it's a seperate function
df.var()
```

#### 1.2 Data Visualization

```python
# almost a direct use of starter code, but expanded to make it more reader friendly
plt_hist = df.hist(figsize=(15,8))
```

```python
# above does not have names, so another graph is needed
feature_names = df.groupby("Name")
feature_names.plot(kind='hist',figsize=(15,8))
```

```python
# below a direct use of the starter code
box = df.boxplot(grid=False, return_type='axes')
```
