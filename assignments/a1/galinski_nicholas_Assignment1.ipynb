This dataset contains 1 million ratings collected from 6000 users on 4000 movies, and it is organized into three tables:
- Ratings
- Users
- Movie information

Each table is available as a separate file, each containing a series of rows where columns are separated by ::

##### Starting code from the link above

```python
import pandas as pd
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('ml-1m/users.dat', sep='::', header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ml-1m/ratings.dat', sep='::', header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('ml-1m/movies.dat', sep='::', header=None, names=mnames, engine='python')
```

```python
users[:5]
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data.head(1)
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:5]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles[:5]
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings[:5]
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

```python
top_male_ratings = mean_ratings.sort_values(by='M', ascending=False)
top_male_ratings[:10]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:10]
```

```python
sorted_by_diff[::-1][:10]
```

###### An aggregate of the movie ratings for each particular genre

```python
genre_ratings = pd.pivot_table(data,index=["genres"], values=["rating"])
genre_ratings[:5]
```

###### The top five highest rated genres by women

```python
avg_genre = data.pivot_table('rating', index='genres', columns='gender', aggfunc='mean')
top_female_genre = avg_genre.sort_values(by='F', ascending=False)
top_female_genre[:5]
```

###### The top five highest rated genres by men

```python
top_male_genre = avg_genre.sort_values(by='M', ascending=False)
top_male_genre[:5]
```

###### The Lion King (1994), broken down by age ranges

###### A function that takes in a user_id and a movie_id, and returns a list of other user_id that rated the movie_id the same rating

```python
def users_with_same_rating(user, movie):
    rating = ratings[(ratings['user_id'] == user) & (ratings['movie_id'] == movie)]['rating'].values[0]
    users_list = list(ratings[(ratings['movie_id'] == movie) & (ratings['rating']==rating)]['user_id'])
    users_list.remove(user)
    return users_list

# print the users with that have the same rating of movie 364 as user 1
print (users_with_same_rating(1,364))
```
